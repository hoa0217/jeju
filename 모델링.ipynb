```python
import pandas as pd #DataFrame
import numpy as np #수식계산
# import seaborn as sns
import matplotlib.pyplot as plt #시각화
from sklearn import tree #결정트리
from sklearn.neighbors import KNeighborsClassifier #KNN 분류,회귀 모델
from sklearn.ensemble import RandomForestClassifier #RandomForest
from sklearn.svm import SVC #서포트벡터
from sklearn.linear_model import SGDClassifier, LogisticRegression #확률적경사하강법, 시그모이드나 서프트벡터 적용
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc #혼돈행렬
from sklearn.preprocessing import StandardScaler, LabelEncoder #평균을 0, 분산을 1 / 
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score #train,test나누기/ 그리드서치/교차검증
%matplotlib inline
```

```python
total = pd.read_csv('total.csv',encoding='euc=kr')
```

```python
total
```

```python
total.info()
```

```python
total['recharge_shop'].unique() 
```

```python
total['recharge_shop'].value_counts()
```

```python
#범주6개
bins = (-1,5,10,15,20,25,31)
group_names = [1,2,3,4,5,6]
total['recharge'] = pd.cut(total['recharge_shop'], bins = bins, labels = group_names)
```

```python
total
```

```python
X = total.drop(['자치센터명','recharge_shop','recharge'], axis = 1) 
y = total['recharge'] 
```

```python
X.head(5)
```

```python
y.value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 7)
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

### KNN

```python
clf = KNeighborsClassifier(n_neighbors=15)
clf = clf.fit(X_train, y_train)
clf.score(X_test,y_test)
```

### 경사하강법

```python
sgd = SGDClassifier()
sgd.fit(X_train, y_train)
sgd.score(X_test,y_test)
```

### 결정트리

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
clf.score(X_test,y_test)
```

### 랜덤 포래스트 (Random Forest Classifier)

```python
rfc = RandomForestClassifier(n_estimators=5000,max_depth=2) 
rfc.fit(X_train, y_train)
print("훈련 세트 정확도: {:.3f}".format(rfc.score(X_train,y_train)))
print("훈련 세트 정확도: {:.3f}".format(rfc.score(X_test,y_test)))
```

### 그라디언트부스팅

```python
from sklearn.ensemble import GradientBoostingClassifier


#n_estimators=100, max_depth=3
gbrt = GradientBoostingClassifier(n_estimators=250,max_depth=1,random_state=0,learning_rate=0.01)
gbrt.fit(X_train, y_train)

print("훈련 세트 정확도: {:.3f}".format(gbrt.score(X_train, y_train)))
print("테스트 세트 정확도: {:.3f}".format(gbrt.score(X_test, y_test)))
```

## Support Vector Classifier (SVC)

```python
svc = SVC()
svc.fit(X_train, y_train)
svc.score(X_test,y_test)
```

# XGBoost

```python
import xgboost as xgb
```

```python
model = xgb.XGBRegressor(booster='gbtree',
                         objective = 'reg:linear',
                           silent=True,
                          min_child_weight=10, 
                            n_estimators = 100,
                         learning_rate = 0.2,
                          n_jobs =4,
                          gamma=0,
                         max_depth = 10,
                         max_delta_step=1,
                         nthread=-1,
                         seed=31,
                         colsample_bytree=0.6,
                         colsample_bylevel=0.9)
model.fit(X_train,y_train,eval_set=[(X_test,y_test)])
```

```python
model.score(X_train,y_train)
```
