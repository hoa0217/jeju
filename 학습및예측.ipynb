```python
import pandas as pd #DataFrame
import numpy as np
```

```python
train= pd.read_csv('total.csv',encoding='euc-kr')
test= pd.read_csv('total_p.csv',encoding='euc-kr')
```

```python
train
```

```python
test
```

```python
#범주6개
bins = (-1,5,10,15,20,25,31)
group_names = [1,2,3,4,5,6]
train['recharge'] = pd.cut(train['recharge_shop'], bins = bins, labels = group_names)
```

```python
train
```

```python
X_train= train.drop(['자치센터명','recharge_shop','recharge'], axis = 1) 
y_train= train['recharge'] 
```

```python
X_test= test.drop(['자치센터명','recharge'], axis = 1) 
y_test= test['recharge']
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

```python
X_test
```

# XGBoost

```python
import xgboost as xgb
```

```python
model = xgb.XGBRegressor(booster='gbtree',
                         objective = 'reg:linear',
                           silent=True,
                          min_child_weight=10, 
                            n_estimators = 100,
                         learning_rate = 0.2,
                          n_jobs =4,
                          gamma=0,
                         max_depth = 10,
                         max_delta_step=1,
                         nthread=-1,
                         seed=31,
                         colsample_bytree=0.6,
                         colsample_bylevel=0.9)
model.fit(X_train,y_train)
```

```python
prediction = model.predict(X_test)
prediction
```

```python
test['recharge']=prediction
```

```python
test
```
